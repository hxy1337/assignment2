```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df6=pd.read_csv('https://raw.githubusercontent.com/hxy1337/assignment2/refs/heads/main/leeds_car_msoa.csv')
df6
```

```python
df7=pd.read_csv('https://raw.githubusercontent.com/hxy1337/assignment2/refs/heads/main/leeds_economic_msoa.csv')
df7
```

```python
df8=pd.merge(df6,df7,on='mnemonic',how='outer')
df8
```

```python
geojson_url2 = 'https://raw.githubusercontent.com/hxy1337/assignment2/main/Leeds_msoa_FeaturesToJSON.geojson'
gdf2 = gpd.read_file(geojson_url2)
gdf2
```

```python
# å‡è®¾ CSV æ–‡ä»¶çš„å­—æ®µåæ˜¯ 'csv_area_code'ï¼ŒGeoJSON æ–‡ä»¶çš„å­—æ®µåæ˜¯ 'geo_area_id'
merged_data = gdf2.merge(df8, left_on='MSOA21CD', right_on='mnemonic', how='left')
merged_data.columns
```

```python
out1= merged_data.copy()# copy a new 'countBusStopLeedsWards' and call 'busStopLeedsWardsCentroid'
out1['geometry'] = out1['geometry'].centroid #Calculates the centroid of each geometry

out1.head()# check the new data 'countBusStopLeedsWards'
```

```python
out1=out1.to_crs(epsg=3857)
merged_data=merged_data.to_crs(epsg=3857)
```

```python
# Data Visualisation
fig, ax = plt.subplots(figsize=(15, 15)) # set the size of subplots

layer1 =merged_data.plot( # define first layer and column base on variable:'bus, minibus or coach'
    ax = ax,
    column = 'Economically active (excluding full-time students):In employment',
    legend = False,  #do not show the legend automatically
    cmap = 'CMRmap',  #select a suitable color
    linewidth = 0.8, #set boundaries for each ward and adjust its line width
    edgecolor = (0, 0, 0, 0.5), #prompt: learned from AI ChatGPT-3.5: The fourth parameter adjusts the opacity of boundaries.
)

layer2 = out1.plot( #define second layer and column base on bus stop numbers in each wards
    ax = ax,
markersize = out1['No cars or vans in household'],
    column = 'No cars or vans in household',
    cmap = 'Oranges', #select a suitable color
    edgecolor = (0, 0, 0, 0.5),
    legend = True, #generate a automatic legend
    k = 5, #classify the legend to 5 class base on 'natural_breaks' method.
    scheme = 'natural_breaks',
)
```

```python
# import required packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Data-preprocessing
df = pd.read_csv('https://raw.githubusercontent.com/FrancescaPontin/GEOG5990M/main/data/week_18/OA_GEOG990M_data.csv.zip',compression='zip', )
```

```python
# check the info of data and whether it contains null value
df.info()
```

```python
#explore the data
df.columns
```

```python
df.head()
```

```python
#select relevant variables and clear the data
#1.make it easy to be operated(to lower the str of columns)
df.columns=df.columns.str.lower()
df.columns
```

```python
#2.select the data(study the correlaiton between the population and the percentage of employment in Manchester region base on the 'consumer vulnerability cluste' variable)
dataSelect=df[['population','economically_active_in_employment_perecentage','consumer_vulnerability_cluster','sub_region']] # select the 4 columns that i need

dataSelect=dataSelect.loc[dataSelect['sub_region']=='Manchester'] # select the target region (Manchester) and exclude another region, two '=' used to check if two values are equal, and return a bool

dataSelect.head() # check the data again
```

```python
#3 check and clear the unreasonable data(population < 0 or percentage > 100)
dataSelect = dataSelect.loc[(dataSelect['population'] > 0) & (dataSelect['economically_active_in_employment_perecentage'] <= 100)]

#4 due to 'consumer_vulnerability_cluste':c represent Students and Young Professionals, student usually do not have job, so we i need to exclude this class
dataSelect = dataSelect[dataSelect["consumer_vulnerability_cluster"] != "C"] # Prompt:learned from use AI: chatgpt3-5 to search the 'unequal to' operator '!='

dataSelect
```

```python
# Data visualisation
# Draw a jointplot, cause The main advantage of sns.jointplot is to simultaneously visualize the relationship between two variables and understand their individual distributions.
sns.jointplot(
    data = dataSelect,# select the data
    x="population", # set the x-axis variable
    y="economically_active_in_employment_perecentage",# set the x-axis variable
    hue="consumer_vulnerability_cluster", # set the additional variable
    palette="Set1", # set a apparent palette class to distinguish different Clusters
    alpha=0.8, # transparency about the element in plot
    height=12 # make the plot bigger
)

# set the title,label and legend suitably
plt.title("Relationship between Population and Employment Rate in Manchester", fontsize=18) # set the title and modify the fontsize
plt.xlabel("Population", fontsize=14) # set the x label and modify the fontsize
plt.ylabel("Employment Rate (%)", fontsize=14) # set the y label and modify the fontsize
plt.legend(title="Consumer Vulnerability Cluster") # set the title of legend
plt.grid(True, linestyle="--", alpha=0.7) # Prompt: Learned from AI ChatGPT-3.5, to set a gird for reference(True is a bool, to control the switch of grid, and other parameter for setting the grid style. linestyle for set the linestyle, alpha for setting the transparency)

# make xticks and yticks more clear
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# make the axis thicker
ax = plt.gca() # Prompt: get knowledge and use-pattern from AI ChatGPT-3.5 (first line code for accessing the current axis, 'gca()' is method for 'get current axis')
ax.spines["bottom"].set_linewidth(2.5)# parameter in [] for select the position of axis and set_linewidth is a method to modify the width of axis
ax.spines["left"].set_linewidth(2.5)

plt.show()

```

```python
# import any further required packages
!pip install contextily
!pip install geoplot
!pip install git+https://github.com/pmdscully/geo_northarrow.git
import geopandas as gpd
import geoplot as gplt
import geoplot.crs as gcrs
from geo_northarrow import add_north_arrow
import contextily as ctx
import matplotlib as mpl
```

```python
# Data-preprocessing

#This block, i choose the 'West_Yorkshire_bus_stops.geojson' and 'leeds_travel_to_work_mode_distance.geojson' this two data which from module 6
bus_stops =gpd.read_file('https://github.com/FrancescaPontin/GEOG5990M/raw/refs/heads/main/data/week_6_7/West_Yorkshire_bus_stops.geojson')
leeds=gpd.read_file('https://github.com/FrancescaPontin/GEOG5990M/raw/refs/heads/main/data/week_6_7/leeds_travel_to_work_mode_distance.geojson')
```

```python
bus_stops.info()# check the info of data and whether it contains null.
```

```python
leeds.head()
```

```python
bus_stops.head()
```

```python
# make it easy to be operated(to lower the str of columns)
bus_stops.columns=bus_stops.columns.str.lower()
leeds.columns=leeds.columns.str.lower()
```

```python
bus_stops['count']=1 #create a new columns for counting the amount of bus stops

bus_stops.columns #check information of columns
```

```python
leeds.columns #check information of columns
```

```python
#clear and select data
countBusPopleedsWards=leeds[['wd24nm','bus, minibus or coach','geometry']].dissolve(by='wd24nm',aggfunc='sum')#create a new data called 'countBusPopleedsWards' for Removing untarget columns and selecting 'bus, minibus or coach' as a variable base on 'wd24nm'

countBusPopleedsWards.head() # check the new data 1
```

```python
bus_stops_leeds=gpd.sjoin(leeds,bus_stops,predicate='intersects') #create a new data called 'bus_stops_leeds' for selecting all the bus stops within Leeds area.

countBusStopLeedsWards=bus_stops_leeds[['wd24nm','count','geometry']].dissolve(by='wd24nm',aggfunc='sum')# Filter out the other columns and remain the necessary columns, calculate the number of bus stops within each 'wd24nm'.

countBusStopLeedsWards.head()# check the new data 'countBusStopLeedsWards'
```

```python
busStopLeedsWardsCentroid = countBusStopLeedsWards.copy()# copy a new 'countBusStopLeedsWards' and call 'busStopLeedsWardsCentroid'

busStopLeedsWardsCentroid['geometry'] = busStopLeedsWardsCentroid['geometry'].centroid #Calculates the centroid of each geometry

busStopLeedsWardsCentroid.head()# check the new data 'countBusStopLeedsWards'
```

```python
countBusPopleedsWards.crs # show the crs
```

```python
busStopLeedsWardsCentroid.crs # show the crs
```

```python
#switch the crs of these two data from 'epsg:27700' to 'epsg:3857'
countBusPopleedsWardsTrans=countBusPopleedsWards.to_crs(epsg=3857)
busStopLeedsWardsCentroidTrans=busStopLeedsWardsCentroid.to_crs(epsg=3857)
```

```python
# Data Visualisation
fig, ax = plt.subplots(figsize=(15, 15)) # set the size of subplots

layer1 = countBusPopleedsWardsTrans.plot( # define first layer and column base on variable:'bus, minibus or coach'
    ax = ax,
    column = 'bus, minibus or coach',
    legend = False,  #do not show the legend automatically
    cmap = 'PuBuGn',  #select a suitable color
    linewidth = 0.8, #set boundaries for each ward and adjust its line width
    edgecolor = (0, 0, 0, 0.5), #prompt: learned from AI ChatGPT-3.5: The fourth parameter adjusts the opacity of boundaries.
)

layer2 = busStopLeedsWardsCentroidTrans.plot( #define second layer and column base on bus stop numbers in each wards
    ax = ax,
    markersize = busStopLeedsWardsCentroidTrans['count'], # adjust the marksize base on the number of bus stops
    column = 'count',
    cmap = 'Oranges', #select a suitable color
    edgecolor = (0, 0, 0, 0.5),
    legend = True, #generate a automatic legend
    k = 5, #classify the legend to 5 class base on 'natural_breaks' method.
    scheme = 'natural_breaks',
)

# create the legend for layer1 manually
lengend1COlor = plt.cm.ScalarMappable( #Prompt: Learned from AI ChatGPT-3.5, to create a new 'scalarMappable' which as the color of layer1
    cmap = 'PuBuGn',
    # Normalize the color scale range; vmin and vmax represent the minimum and maximum values of the data.
    norm = mpl.colors.Normalize(
        vmin = countBusPopleedsWardsTrans['bus, minibus or coach'].min(),
        vmax = countBusPopleedsWardsTrans['bus, minibus or coach'].max())
)
# prompt: Learned from AI, to draw a colorbar as the legend for first layer
legend1 = fig.colorbar(
    lengend1COlor, # choose above scalarMappable as the range of color
    ax = ax,
    fraction = 0.035, #the width of the bar
    pad = 0.0 #the distance between the main plot and colorbar
    )

legend1.set_label( #give a title for legend1
    'Number of people using bus, minibus or coach',
    fontsize = 12 #control the font size
    )

#adjust the legend 2 appropriately
legend2 = layer2.get_legend()
legend2.set_title('Bus Stops Count') #set a title
legend2.get_title().set_fontsize(12) #adjust text size
legend2.set_bbox_to_anchor((0.02, 0.02)) #adjust the position of legend 2

#prompt: learned from AI ChatGPT-3.5: set a frame for legend 2
legend2.set_frame_on(True)
legend2.get_frame().set_edgecolor('black') # set the frame color to black
legend2.get_frame().set_facecolor('lightblue') # change the background color to lightblue

#set a background map
ctx.add_basemap(
    ax,
    source=ctx.providers.CartoDB.Voyager,
    zoom = 13,
    crs = countBusPopleedsWardsTrans.crs
    )

# set a main title fron whole plot
ax.set_title('Distribution of Bus Usage and Number Of Bus Stop Across Leeds Wards', fontsize=20)

#set a north arrow
add_north_arrow(
    ax = ax,
    scale = .75, # the size of Arrow
    # the position of north arrow in this map
    xlim_pos = 0.1,
    ylim_pos = .85,
    color = '#000',
    text_scaler = 2, # the "N" size
    text_yT = 6.25) # adjust the "N" to suitable position

ax.set_axis_off() # close the axis for good looking
```

In this spatial data visualization, I mainly focus on representing the distribution of people using buses, minibuses, or coaches across Leeds wards, while incorporating the number of bus stops as an additional variable to provide further information. For second legend, I modified some parameters to adjust its style and position. The frame was show and the background color was set to light blue to avoid misleading color.

A CartoDB Voyager basemap was added to provide geographic reference, and a north arrow aids orientation. The axis lines were removed to keep the map clean and visually focused.
